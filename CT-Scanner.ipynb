## Algorytm Bresenhama
Na podstawie algorytmu https://eduinf.waw.pl/inf/utils/002_roz/2008_06.php

```python
def bresenham(x1, y1, x2, y2):
    kx = 1 if x2 >= x1 else -1
    ky = 1 if y2 >= y1 else -1

    dx = abs(x2 - x1)
    dy = abs(y2 - y1)

    pixels = [(x1, y1)]
    if dx < dy:
        e = dy / 2
        for _ in range(dy):
            y1 += ky
            e -= dx
            if e < 0:
                x1 += kx
                e += dy
            pixels.append((x1, y1))
    else:
        e = dx / 2
        for _ in range(dx):
            x1 += kx
            e -= dy
            if e < 0:
                y1 += ky
                e += dx
            pixels.append((x1, y1))

    return pixels

print(bresenham(1, 1, 7, 5))
```

TESTOWA KOMÓRKA

```python
def bresenhamGenerator(x0, y0, x1, y1):
    dx = x1 - x0
    dy = y1 - y0

    xsign = 1 if dx > 0 else -1
    ysign = 1 if dy > 0 else -1

    dx = abs(dx)
    dy = abs(dy)

    if dx > dy:
        xx, xy, yx, yy = xsign, 0, 0, ysign
    else:
        dx, dy = dy, dx
        xx, xy, yx, yy = 0, ysign, xsign, 0

    D = 2*dy - dx
    y = 0

    for x in range(dx + 1):
        yield [x0 + x*xx + y*yx, y0 + x*xy + y*yy]
        if D >= 0:
            y += 1
            D -= 2*dx
        D += 2*dy

print(list(bresenhamGenerator(1, 1, 7, 5)))
```
